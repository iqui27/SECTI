```python
import requests
from datetime import datetime

# Defina a interface do usurio
origem = input('Origem: ')
destino = input('Destino: ')
data_ida = input('Data da ida (formato: AAAA-MM-DD): ')
data_volta = input('Data da volta (formato: AAAA-MM-DD): ')
one_way = input('Somente ida (s/n): ').lower() == 's'

# Defina os parmetros da consulta
classe_servico = 'ECONOMY'  # classe de servio desejada
adultos = 1  # nmero de adultos
max_preco = 300000  # preo mximo em centavos (R$ 3000)

# Defina o cabealho Authorization
api_key = 'VAsxLeaADuykE9EuESeAz8ZNKGRP'
headers = {
    'Authorization': f'Bearer {api_key}',
    'User-Agent': 'Mozilla/5.0'
}

# Faa a solicitao para a API do Amadeus
url = 'https://test.api.amadeus.com/v2/shopping/flight-offers'
params = {
    'originLocationCode': origem,
    'destinationLocationCode': destino,
    'departureDate': data_ida,
    'returnDate': data_volta if not one_way else None,
    'currencyCode': 'BRL',
    'adults': adultos,
    'max': 50,  # nmero mximo de resultados
    'maxPrice': max_preco,
    'travelClass': classe_servico,
    'nonStop': 'false',  # permitir voos com escalas
}

response = requests.get(url, params=params, headers=headers)
print(response.text)

# Analise os resultados da solicitao
if response.status_code == 200:
    data = response.json()
    # Obtenha os resultados das passagens mais baratas
    print('\nResultados da pesquisa:\n')
    for oferta in data['data']:
        preco = oferta['price']['total']
        ida = oferta['itineraries'][0]['segments'][0]['departure']['at']
        volta = oferta['itineraries'][0]['segments'][-1]['arrival']['at']
        empresa = oferta['validatingAirlineCodes'][0]
        print(f"Preo: R$ {preco}, Data de ida: {ida}, Data de volta: {volta}, Companhia area: {empresa}")
else:
    print(f"No foi possvel obter resultados. Cdigo de status: {response.status_code}")
    print(f"Mensagem de erro: {response.text}")
```
